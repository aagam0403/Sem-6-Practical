```python
#Write a Machine Learning code to perform following:
#• Encode categorical variables appropriately.
#• Train a decision tree classifier.
#• Visualize the decision tree.
#• Predict the activity for:
#• Temperature = Mild, Humidity = Normal, Wind Speed = Strong
```

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
```

```python
data ={
    'Temprature':['Hot','Hot','Mild','Cold','Mild','Cold'],
    'Humidity':['High','High','Normal','Normal','High','High'],
    'Wind Speed':['Weak','Strong','Weak','Strong','Weak','Strong'],
    'Activity':['No','No','Yes','Yes','Yes','NO']}

df=pd.DataFrame(data)
print(df)
    
```

```python
# Encode categorical variables appropriately.
le=LabelEncoder()
for column in df.columns:
    df[column]=le.fit_transform(df[column])

print(df)
```

```python
#Train a decision tree classifier.
X=df[['Temprature','Humidity','Wind Speed']]
Y=df['Activity']

model=DecisionTreeClassifier()
model.fit(X,Y)
```

```python
#Visualize the decision tree
plt.figure(figsize=(10,6))
plot_tree(model,feature_names=['Temprature','Humidity','Wind Speed'], filled=True)
plt.title("Decision Tree Visualization")
plt.show()
```

```python
#Prediction
test_data=[[2,1,0]]
prediction=model.predict(test_data)
print(prediction[0])
```
