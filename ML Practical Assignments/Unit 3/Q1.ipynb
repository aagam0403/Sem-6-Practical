```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
```

```python
data ={
    'Age':[25,30,45,35,50,28],
    'Income':[3,5,8,4,10,3],
    'Credit Score':[650,700,720,680,750,640],
    'Loan Approved':['No','Yes','Yes','No','Yes','No']
}
df=pd.DataFrame(data)
print(df)
    
```

```python
#Convert Loan Approved Categorical data to Numeric Data
le=LabelEncoder()
df['Loan Approved']=le.fit_transform(df['Loan Approved'])
print(df)
```

```python
# Consider Age, Income and Credit Score as an input field and Loan Approved as an output 
X=df[['Age','Income','Credit Score']]
Y=df['Loan Approved']
```

```python
# Train a model using decision tree method.
model= DecisionTreeClassifier()
model.fit(X,Y)
```

```python
# Test a model is Age is 30, Income is 8 Lacs and Credit Score is 700, Whether Loan 
#approved or not.
test_data=[[30,8,700]]
prediction=model.predict(test_data)

result=le.inverse_transform(prediction)
print("Prediction for Age=30,Income=8,Credit Score =700",result[0])
```

```python
plt.scatter(df['Age'],df['Income'],c=df['Loan Approved'])
plt.xlabel("Age")
plt.ylabel("Income")
plt.title("Age vs Income (Loan Approved)")
plt.colorbar(label="Loan Approved (0=No,1=Yes)")
plt.show
```
