```python
#4. Write a Machine Learning program to implement Simple Linear Regression for predicting 
#house prices based on size.
#Size (sq ft) 500 800 1200 1500 1800
#Price ($) 50000 80000 120000 150000 180000
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
size=np.array([500,800,1200,1500,1800]).reshape(-1,1)
price=np.array([50000,80000,120000,150000,180000])
```

```python
#Create model
model=LinearRegression()
```

```python
#Train Model
model.fit(size,price)
```

```python
#predice price for new size
new_size=np.array([[1000]])
predicted_price=model.predict(new_size)
print(predicted_price[0])
```

```python
#Visualization
plt.scatter(size,price)
plt.plot(size,model.predict(size))
plt.xlabel("Size")
plt.ylabel("Price")
plt.title("House Price Prediction")
plt.show()
```
