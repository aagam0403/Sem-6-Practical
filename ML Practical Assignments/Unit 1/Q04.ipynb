```python
#Write a Machine Learning program to filter out irrelevant or erroneous 
#data points from a dataset based on predefined criteria,  
#1. Age between 25 to 60 
#2. Salary greater than 10000 
#data = { 
#'Name': ['John', 'Anna', 'Peter', 'Linda'], 
#'Age': [24, 13, 53, 33],  
#'Salary': [50000, 2000, 100000, 30000] 
```

```python
import pandas as pd
```

```python
data = {
    'Name': ['John','Anna','Peter','Linda'],
    'Age': [24,13,53,33],
    'Salary': [50000,2000,100000,30000]
}

df=pd.DataFrame(data)
print(df)
```

```python
#Apply filtering Conditions
filtered_df=df[
    (df['Age']>=25)&
    (df['Age']<=60)&
    (df['Salary']>10000)
    ]

print("Filtered Data:")
print(filtered_df)
```
