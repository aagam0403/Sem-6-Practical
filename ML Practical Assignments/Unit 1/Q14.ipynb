```python
import pandas as pd
```

```python
#1. Create a DataFrame with the following data: 
#o Name: ['Alex', 'Bob', 'Clarke'] 
#o Age: [10, 12, 13] 
#o Print the DataFrame. 

data1 = {
    'Name': ['Alex', 'Bob', 'Clarke'],
    'Age': [10, 12, 13]
}
df1 = pd.DataFrame(data1)

print("1. Created DataFrame:")
print(df1)
```

```python
#2. Read a CSV file named employees_info.csv and display its 
#contents
df=pd.read_csv('employees_info.csv')

print("CSV Data:")
print(df)
```

```python
#3. Get the general information of the DataFrame (such as column 
#names, data types, and memory usage).
print("Info")
print(df.info())
```

```python
# 4. Access columns\
print ("Name and Gender")
print(df[['name','gender']])
```

```python
# 5. First row using loc
print("First Row")
print(df.loc[0])
```

```python
# 6. Rows 0–5 with specific columns
print(df.loc[0:5, ['name','job title']])
```

```python
# 7. Filter Accounting department
print(df[df['department']=='Accounting'][['name','job title','department']])
```

```python
# 8. Delete 'time zone' column
df=df.drop(columns=['time zone'],errors='ignore')

print(df)
```

```python
# 9. Drop duplicates
print(df.drop_duplicates())
```

```python
# 10. Drop duplicates based on 'residence'
print(df.drop_duplicates(subset='residence'))
```

```python
# 11. Drop rows with missing values
print(df.dropna())
```

```python
# 12. Drop columns with missing values
print(df.dropna(axis=1))
```

```python
# 13. Threshold (at least 2 non-NaN)
print(df.dropna(thresh=2))
```

```python
# 14. Count missing values
print(df.isnull().sum())
```

```python
# 15. Percentage of missing values
print((df.isnull().sum()/len(df)) * 100)
```

```python
# 16. Fill missing values
print(df.fillna("Unknown"))
```

```python
# 17. Standardize 'name' column
df['name_title']=df['name'].str.title()
df['name_lower']=df['name'].str.lower()

print(df[['name', 'name_title', 'name_lower']])
```

```python
# 18. Replace gender values
# -------------------------------
df['gender'] = df['gender'].replace({'M': 'Male', 'F': 'Female'})

print("\n18. Updated Gender:")
print(df['gender'])
```

```python
# 19. Clean phone column (remove hyphens)
# -------------------------------
df['phone'] = df['phone'].astype(str).str.replace('-', '', regex=True)

print("\n19. Cleaned Phone Numbers:")
print(df['phone'])
```
