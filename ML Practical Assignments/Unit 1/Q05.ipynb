```python
#Write a Machine Learning program to clean textual data by removing 
#HTML tags, special characters, and punctuation. Use Python's re library 
#to demonstrate this process. 
#text = "<html>Hello! This is <b>clean</b> text.</html>" 
```

```python
import re
```

```python
text="<html>Hello! This is <b>clean</b> text.</html>"
print("Original Text:")
print (text)
```

```python
#Remove HTML Tags
clean_text=re.sub(r'<.*?>', '',text)

#Remove Special Characters and punctuation
clean_text=re.sub(r'[^a-zA-Z0-9\s]', '', clean_text)

#Convert multiple space into single space

clean_text=re.sub(r'\s+', ' ',clean_text).strip()

print("Cleaned Text")
print(clean_text)
```
