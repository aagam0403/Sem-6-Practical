```python
import pandas as pd
```

```python
data={
    'Name':['John','Anna','Peter','Linda'],
    'Location':['New York','Paris','Berlin','London'],
    'Age':[24,13,53,33]
}

df=pd.DataFrame(data)
```

```python
#1. Display the entire DataFrame.
print(df)
```

```python
#2. Select and display all rows where the age is greater than 30. 
print("Age > 30:")
print(df[df['Age'] > 30])
```

```python
#3. Display the details of the first person (row with index 0). 
print("First Person")
print(df.loc[0])
```

```python
#4. Display the details of the first two people (rows with indexes 0 
#and 1). 
print("First two person:")
print(df.loc[0:1])
```
