```python
#Write a Machine Learning program to scale numerical features in a 
#dataset using Min-Max scaling.  
#data = {'Income': [15000, 1800, 120000, 10000], 'Age': [25, 18, 42, 
#51]} 
```

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
```

```python
data={
    'Income': [15000,1800,120000,10000],
    'Age': [25,18,42,51]}

df=pd.DataFrame(data)
print(df)
```

```python
#Initalize Scaler
scaler=MinMaxScaler()
```

```python
#Apply Scaling
df_scaled=pd.DataFrame(scaler.fit_transform(df),columns=df.columns)

print("Scaled Data")
print(df_scaled)
```
