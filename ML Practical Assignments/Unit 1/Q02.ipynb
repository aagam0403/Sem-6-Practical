```python
#Write a Machine Learning program to handle missing values in a 
#dataset. Demonstrate two approaches: 
#● Deleting rows or columns with missing values using dropna(). 
#● Imputing missing values using strategies like mean, median, or a 
#specified constant. 
#data = {'Name': ['John', 'Anna', 'Peter', None], 'Age': [24, 13, 
#None, 33]} 
```

```python
import pandas as pd
from sklearn.impute import SimpleImputer
```

```python
#Given data
data = {
    'Name': ['John','Anna','Peter',None],
    'Age': [24,13,None,33]
}

df = pd.DataFrame(data)
print(df)
```

```python
#Deleting missing values
#Drop rows with missing values

df_drop_rows=df.dropna()

print("\n After dropping rows with missing values:")
print(df_drop_rows)
```

```python
#Drop column with missing values
df_drop_column=df.dropna(axis=1)

print("After dropping column with missing values")
print(df_drop_column)
```

```python
data = {
    'Name': ['John','Anna','Peter',None],
    'Age': [24,13,None,33]
}

df = pd.DataFrame(data)
print(df)

```

```python
# Mean Imputation (for numerical column 'Age')
mean_imputer = SimpleImputer(strategy='mean')
df['Age_mean'] = mean_imputer.fit_transform(df[['Age']]).ravel()

# Median Imputation
median_imputer = SimpleImputer(strategy='median')
df['Age_median'] = median_imputer.fit_transform(df[['Age']]).ravel()

# Constant Imputation (for Name column)
```

```python

# Constant Imputation (for Name column)
const_imputer = SimpleImputer(strategy='constant', fill_value='Unknown')
df['Name_filled'] = const_imputer.fit_transform(df[['Name']]).ravel()

print("\nAfter Imputation:")
print(df)
```
