```python
#Write a Machine Learning program to convert categorical variables into 
#numerical representations using one-hot encoding and label encoding 
#techniques. Use pandas and sklearn to demonstrate the encoding 
#process. 
#data ={'Department': ['HR', 'Legal', 'Marketing', 'Management']} 
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
data={
    'Department':['HR','Legal','Marketing','Management']}

df=pd.DataFrame(data)
print(df)
```

```python
#Label Encoding

label_encoder=LabelEncoder()
df['Label_Encoded']=label_encoder.fit_transform(df['Department'])

print("After Label Encoding")
print(df)
```

```python
# One-Hot Encoding
one_hot=pd.get_dummies(df['Department'])

print("One-Hot Encoded Data:")
print(one_hot)
```
