```python
#Write a Machine Learning program to standardize inconsistent date 
#formats in a dataset using the to_datetime() method in pandas. 
#data = {'Date': ['2023-01-01', '01/02/2023', '2023.03.03']}
```

```python
import pandas as pd
```

```python
#Given data 
data={
    'Date':['2023-01-01','01/02/2023','2023.03.03']
}
df=pd.DataFrame(data)
print(df)
```

```python
#Convert to Standard datetiem formate
df['Standard_Date']=pd.to_datetime(df['Date'], dayfirst=True, errors='coerce')

print("After Standardizing Dates:")
print(df)
```

```python
#Convert to Uniform format
df['Formatted_Date']= df['Standard_Date'].dt.strftime('%Y-%m-%d')

print("Final Formatted Dates:")
print (df)
```
