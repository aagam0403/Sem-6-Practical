```python
#Write a Machine Learning program to remove duplicate entries from a customer database using the drop_duplicates() method in pandas. 
#Demonstrate how to remove duplicates based on specific columns, keep 
#either the first or last occurrence. 
#data = { 
#'Name': ['John', 'Anna', 'Peter', 'John'], 
#'Age': [24, 13, 53, 24] 
#}
```

```python
import pandas as pd
```

```python
#Given data
data = {
    'Name': ['John','Anna','Peter','John'],
    'Age': [24,13,53,24]
}

#create Dataframe
df=pd.DataFrame(data)

print("Original DataFrame:")
print(df)
```

```python
#Remove duplicate (keep first occurrence)
df_first= df.drop_duplicates()

print("\n After removing duplicate (keep first):")
print(df_first)
```

```python
# Remove duplicates (keep last occurrence)
df_last=df.drop_duplicates(keep='last')

print("After removing duplicate (keep last)")
print(df_last)
```

```python
# Remove duplicated based on specific column (Name)
df_name= df.drop_duplicates(subset='Name',keep='first')

print("Remove based on 'Name' Column")
print(df_name)
```
