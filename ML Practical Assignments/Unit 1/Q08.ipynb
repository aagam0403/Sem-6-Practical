```python
#Write a Machine Learning program to transform skewed distributions 
#using log or square root transformations. Visualize the effect of these 
#transformations using matplotlib. 
#Define data as: 
#data = np.random.exponential(scale=2, size=1000)
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=np.random.exponential(scale=2, size=1000)
```

```python
#Apply transformations
log_data=np.log1p(data)
sqrt_data=np.sqrt(data)
```

```python
#Visualization
plt.figure()
plt.hist(data, bins=30)
plt.title("Original Data(Skewed)")
plt.xlabel("Value")
plt.ylabel("frequency")
plt.show()
```

```python
#Log transformed Data
plt.figure()
plt.hist(log_data,bins=30)
plt.title("Log Transformed Data")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
#Square Root Transformed data
plt.figure()
plt.hist(sqrt_data,bins=30)
plt.title("Square Root Transfromed Data")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```
